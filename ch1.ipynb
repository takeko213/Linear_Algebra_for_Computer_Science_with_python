```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

```python
# たくさんの点x
X_ini = list(itertools.product([i / 10 for i in range(-10, 11)] , repeat=2))
```

```python
# グラフ作成用関数
def make_graph(Y,axlim):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    Y = np.array(Y)
    
    # 点xを描画(散布図)
    ax.scatter(x=Y[:,0], y=Y[:,1], s=10, c='gray')
    ax.set_xlim([-axlim, axlim])
    ax.set_ylim([-axlim, axlim])
    
    # 基底ベクトルを描画（矢印）
    ax.annotate('',xy=Y[230], xytext=(0,0),arrowprops=dict(facecolor='b'))
    ax.annotate('',xy=Y[430], xytext=(0,0),arrowprops=dict(facecolor='g'))
    
    plt.show()

# 写像用関数
def mapping(A,Xs):
    Y = []
    for X in Xs:
        a = np.dot(A, np.array(X))
        Y.append(a)
    return Y
```

# 1.2.3 写像
$\boldsymbol{y} = \boldsymbol{A}\boldsymbol{x}$
<br>
$\boldsymbol{A} = \begin{bmatrix}1 & -0.3 \\ -0.7 & 0.6 \end{bmatrix}$  

```python
# 初期状態
make_graph(Y = X_ini, axlim = 1.1)
```

```python
# 写像 
A = np.array([[1,-0.3],[-0.7,0.6]])
Y = mapping(A,X_ini)
make_graph(Y,axlim = 1.1)
```

# 1.2.4 写像の合成

```python
A = np.array([[0,-1],[1,0]])
B = np.array([[2,0],[0,1]])
```

```python
AB = np.dot(A,B)
AB
```

```python
BA = np.dot(B,A)
BA
```

```python
# AしてB
Y = mapping(A,X_ini)
make_graph(Y, axlim = 2.2)
Y = mapping(B,Y)
make_graph(Y, axlim = 2.2)
```

```python
# BA
Y = mapping(BA,X_ini)
make_graph(Y, axlim = 2.2)
```

```python
# BしてA
Y = mapping(B,X_ini)
make_graph(Y, axlim = 2.2)
Y = mapping(A,Y)
make_graph(Y, axlim = 2.2)
```

```python
# AB
Y = mapping(AB,X_ini)
make_graph(Y, axlim = 2.2)
```

# 1.2.6 行列のべき乗
$(\boldsymbol{A}\boldsymbol{B})^2 = \boldsymbol{A}\boldsymbol{B}\boldsymbol{A}\boldsymbol{B}$

```python
A = np.array([[1,0],[0,0]])
B = np.array([[0,-1],[1,0]])
```

```python
# (AB)^2 = BABA
Y = mapping(B,X_ini)
make_graph(Y, axlim = 1.1)
Y = mapping(A,Y)
make_graph(Y, axlim = 1.1)
Y = mapping(B,Y)
make_graph(Y, axlim = 1.1)
Y = mapping(A,Y)
make_graph(Y, axlim = 1.1)
```

```python
# A^2 B^2 = BBAA
Y = mapping(B,X_ini)
make_graph(Y, axlim = 1.1)
Y = mapping(B,Y)
make_graph(Y, axlim = 1.1)
Y = mapping(A,Y)
make_graph(Y, axlim = 1.1)
Y = mapping(A,Y)
make_graph(Y, axlim = 1.1)
```

# 1.2.7 ゼロ行列・単位行列・対角行列

```python
# ゼロ行列
A = np.array([[0,0],[0,0]])
Y = mapping(A,X_ini)
make_graph(Y, axlim = 1.1)
```

```python
# 単位行列
A = np.array([[1,0],[0,1]])
Y = mapping(A,X_ini)
make_graph(Y, axlim = 1.1)
```

```python
# 全成分が1の行列
A = np.array([[1,1],[1,1]])
Y = mapping(A,X_ini)
make_graph(Y, axlim = 1.1)
```

```python
# 対角行列
A = np.array([[1.5,0],[0,0.5]])
Y = mapping(A,X_ini)
make_graph(Y, axlim = 1.7)
```

```python
# 対角行列（対角成分に0がある例）
A = np.array([[0,0],[0,0.5]])
Y = mapping(A,X_ini)
make_graph(Y, axlim = 1.1)
```

```python
# 対角行列（対角成分に負の数がある例）
A = np.array([[1.5,0],[0,-0.5]])
Y = mapping(A,X_ini)
make_graph(Y, axlim = 1.7)
```
